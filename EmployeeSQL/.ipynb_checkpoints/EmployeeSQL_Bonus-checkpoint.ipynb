```python
# import dependecies
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# import keys
from config import config
```

```python
# create an engine that can talk to the database
engine = create_engine('postgresql://{username}:{password}@localhost:5432/Pewlett_Hackard_employees_db')
connection = engine.connect()
```

```python
# create dataframe for employees, titles and salaries
df_employees =pd.read_sql("select * from employees", connection)
df_titles = pd.read_sql("select * from titles", connection)
df_salaries = pd.read_sql("select * from salaries", connection)
```

```python
# set index of df_salaries in order to create histogram
df_salaries2 = df_salaries.set_index('emp_no')
df_salaries2
```

```python
# create a histogram to visualize the most common salary ranges for employees

# determine fig and ax
fig, ax = plt.subplots()

# determine type of plot and format
df_salaries2.plot(kind='hist',
                 figsize=(7,7),
                 ax=ax,
                 color="limegreen",
                 edgecolor="black",
                 title="Salary Distribution");

# rotate xticks
plt.xticks(rotation = 75)

# apply grid
ax.yaxis.grid()

# display histogram
fig.tight_layout();

# save figure
plt.savefig("../sql-challenge/Images/Bonus - Histogram Salary Distribution.png")
```

```python
# merge dataframes in order to create bar chart of average salary
# merge employees and titles dataframe
df_merge = pd.merge(df_employees, df_titles, left_on='emp_title_id', right_on='title_id', how='left')
df_merge
```

```python
# merge with salaries
df_merge = pd.merge(df_merge, df_salaries, on='emp_no', how='left')
df_merge
```

```python
# create dataframe for average salary by title
df_average_salary = df_merge.groupby('title')['salary'].mean().reset_index()

# set index to column "title"
df_average_salary = df_average_salary.set_index('title')

# sort values in descending order
df_average_salary = df_average_salary.sort_values(by=["salary"], ascending=False)

# display dataframe
df_average_salary
```

```python
# create a bar chart of average salary by title

# determine fig and ax
fig1, ax1 = plt.subplots()

# determine type of plot and format
df_average_salary.plot(kind="bar",
                       figsize=(7,7),
                       ax=ax1,
                       color="limegreen",
                       edgecolor="black",
                       xlabel="Job Title",
                       ylabel="Average Salary ($)",
                       title="Average Salary by Job Title");

# rotate xticks
plt.xticks(rotation = 75)

# apply grid
ax1.yaxis.grid()

# display bar chart
fig1.tight_layout();

# save figure
plt.savefig("../sql-challenge/Images/Bonus - Bar Chart Average Salary.png")
```
